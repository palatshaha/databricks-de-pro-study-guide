Exam focus: premature optimization and cost explosion.

## Context
Z-Ordering is often misused as a default optimization.

```

from pyspark.sql.functions import rand
df = spark.range(0, 500_000).withColumn("group_id", (rand()*100).cast("int"))
df.write.mode("overwrite").saveAsTable("de_pro_week1.write_heavy")
```

```

OPTIMIZE de_pro_week1.write_heavy ZORDER BY (group_id)
```

## Exam Takeaway
Optimization must follow stable access patterns.

Exam focus: understanding Delta MERGE cost.

```

from pyspark.sql.functions import rand
base = spark.range(0, 100_000).withColumn("value", rand())
base.write.mode("overwrite").saveAsTable("de_pro_week1.silver_target")
```

```

updates = spark.range(0, 10_000).withColumn("value", rand())
updates.createOrReplaceTempView("updates")
```

```

MERGE INTO de_pro_week1.silver_target t
USING updates s
ON t.id = s.id
WHEN MATCHED THEN UPDATE SET *
```

## Exam Takeaway
MERGE rewrites files, not rows.

Focus: file-level rewrite behavior.

```

from pyspark.sql.functions import rand
spark.range(0, 100_000).withColumn("v", rand())     .write.mode("overwrite").saveAsTable("de_pro_week2.merge_target")
```

```

updates = spark.range(0, 5000).withColumn("v", rand())
updates.createOrReplaceTempView("updates")
```

```

MERGE INTO de_pro_week2.merge_target t
USING updates s
ON t.id = s.id
WHEN MATCHED THEN UPDATE SET *
```

## Exam Takeaway
MERGE rewrites files, not rows.

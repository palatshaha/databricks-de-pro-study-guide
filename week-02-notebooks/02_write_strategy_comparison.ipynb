# Write Strategy Comparison

```

spark.range(0, 100_000).withColumn("grp", (rand()*10).cast("int"))     .write.mode("overwrite").partitionBy("grp")     .saveAsTable("de_pro_week2.partitioned")
```

```

INSERT OVERWRITE TABLE de_pro_week2.partitioned PARTITION (grp)
SELECT * FROM de_pro_week2.partitioned
```

## Exam Takeaway
Partition overwrite gives predictable cost.
